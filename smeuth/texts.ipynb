```python
import numpy as np
import matplotlib.pyplot as plt

from ipywidgets import *
from IPython.display import display

%matplotlib inline
import seaborn as sns
```

```python
sns.pairplot(dat)
```

```python
dat.corr()
```

```python
from sklearn.decomposition import PCA
```

```python
pca = PCA().fit(dat)
```

```python
pca.components_
```

```python
pca.explained_variance_ratio_
```

```python
fig, ax = plt.subplots(figsize=(8,6))
ax.plot(range(1, dat.shape[1]+1), pca.explained_variance_ratio_, lw=2)
ax.scatter(range(1, dat.shape[1]+1), pca.explained_variance_ratio_, s=100)
ax.set_title('explained variance of components')
ax.set_xlabel('principal component')
ax.set_ylabel('explained variance')
plt.show()
```

```python
pca1_evec = pca.components_[0]
for weight, per in zip(pca1_evec, dat.columns):
    print(per, weight)
```

```python
pca2_evec = pca.components_[2]
for weight, per in zip(pca2_evec, dat.columns):
    print(per, weight)
```

```python
stats_corr = dat.corr()

mask = np.zeros_like(stats_corr, dtype=np.bool)
mask[np.triu_indices_from(mask)] = True

fig, ax = plt.subplots(figsize=(8,7))
ax = sns.heatmap(stats_corr, mask=mask, cmap='RdBu_r')

ax.set_xticklabels(ax.xaxis.get_ticklabels(), fontsize=14, rotation=45)
ax.set_yticklabels(ax.yaxis.get_ticklabels(), fontsize=14, rotation=0)

plt.show()
```

```python
stats = pd.DataFrame(pca.transform(dat), columns=['PC'+str(i) for i in range(1,6)])
stats.head()
```

```python
fig, ax = plt.subplots(figsize=(10,8))
ax.scatter(dat['Openness'].values, dat['Extraversion'].values, s=0)

for i, txt in enumerate(dat.index.values):
    ax.annotate(txt, (0, 0), (dat['Openness'].values[i], dat['Extraversion'].values[i]),
            arrowprops=dict(arrowstyle='<-', color='black', linewidth=1.5),
            xycoords='data', textcoords='data', fontsize=12, color="black")

ax.set_title('Openness vs. Extraversion')
ax.set_xlabel('Openness')
ax.set_ylabel('Extraversion')
plt.show()
```

```python
fig, ax = plt.subplots(figsize=(10,8))
ax.scatter(stats.PC1.values, stats.PC3.values, s=0)

for i, txt in enumerate(stats.index.values):
    ax.annotate(txt, (0, 0), (stats.PC1.values[i], stats.PC3.values[i]),
            arrowprops=dict(arrowstyle='<-', color='black', linewidth=1.5),
            xycoords='data', textcoords='data', fontsize=12, color="black")

ax.set_title('PC1 (Openness) vs. PC3 (Extraversion)')
ax.set_xlabel('principal component 1 (Openness)')
ax.set_ylabel('principal component 3 (Extraversion)')
plt.show()
```
